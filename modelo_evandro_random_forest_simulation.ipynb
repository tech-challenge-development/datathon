# 🧪 Simulador de Previsão de Risco de Defasagem

```python
import pandas as pd
import joblib
import os

# Caminho do modelo salvo no notebook anterior
caminho_modelo = 'modelo_evandro_random_forest_joblib.joblib'

if os.path.exists(caminho_modelo):
    modelo = joblib.load(caminho_modelo)
    print("✅ Modelo carregado com sucesso!")
else:
    print("⚠️ ATENÇÃO: Execute o notebook 'modelo_evandro_random_forest.ipynb' primeiro para gerar o arquivo do modelo.")
```

### 1. Definição de Dados de Entrada do Aluno
Aqui definimos variáveis que seriam preenchidas pelo usuário em um formulário da tela.

```python
fase = 4.0
inde = 6.2
iaa = 7.5
ieg = 6.0
ips = 5.5
ida = 5.0
ano_ingresso = 2022
ano_referencia = 2024
genero = 'MENINO'
pedra = 'AMETISTA'
instituicao = 'ESCOLA PÚBLICA'

print("Dados coletados para simulação.")
```

### 2. Estruturação no Formato Aceito pelo Pipeline
O pipeline exige um DataFrame com as colunas nomeadas exatamente como no treinamento.

```python
# Criando o DataFrame de linha única
df_entrada = pd.DataFrame([{
    'Fase_Cleansed': float(fase),
    'INDE_Consolidado': float(inde),
    'IAA': float(iaa),
    'IEG': float(ieg),
    'IPS': float(ips),
    'IDA': float(ida),
    'Ano ingresso': int(ano_ingresso),
    'Ano_Referencia': int(ano_referencia),
    'Gênero': genero,
    'Pedra_Consolidada': pedra,
    'Instituição de ensino': instituicao
}])

display(df_entrada)
```

### 3. Execução do Modelo e Regra de Negócio
Agora calculamos a probabilidade e aplicamos os limites: **Alto (>=65%)**, **Moderado (40-64%)** e **Baixo (<40%)**.

```python
# Pegando a probabilidade da classe de risco (índice 1)
probabilidade = modelo.predict_proba(df_entrada)[0][1]
percentual = probabilidade * 100

print("=" * 50)
print(f"📊 RESULTADO DA SIMULAÇÃO")
print(f"Probabilidade de Defasagem: {percentual:.2f}%")
print("=" * 50)

# Aplicação das regras de cores e decisão definidas na apresentação
if percentual >= 65:
    status = "🔴 ALTO - Requer intervenção pedagógica imediata."
elif percentual >= 40:
    status = "🟠 MODERADO - Necessário monitoramento preventivo."
else:
    status = "🟢 BAIXO - Seguir com acompanhamento regular."

print(f"\nNível de Risco Classificado: {status}")
```
